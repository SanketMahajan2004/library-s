```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
c = ["r","y","g","b"]
plt.bar(x,y,color=c)
plt.show
```

```python
x1 =["python","c","c++","java"]
y1=[85,70,60,82]
plt.xlabel("languages",fontsize=15)
plt.ylabel("percentage",fontsize=15)
plt.title("Table",fontsize=20)
c=["g","y","m","b"]
plt.bar(x1,y1,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5,label="popularity")
plt.bar(x1,y1,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5,label="popularity")
plt.legend()
plt.show()
```

```python
import numpy as np
x2 =["python","c","c++","java"]
y2=[85,70,60,82]
z2=[20,30,40,50]
p=np.arange(len(x2))
width=0.2
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=15)
plt.ylabel("percentage",fontsize=15)
plt.title("Table",fontsize=20)

plt.bar(x2,y2,width,color="r",linewidth=5,linestyle=":",label="popularity")
plt.bar(p1,z2,width,color="y",linewidth=5,linestyle=":",label="popularity1")

plt.xticks(p+width/2,x2,rotation=20)
plt.legend()
plt.show()
```
